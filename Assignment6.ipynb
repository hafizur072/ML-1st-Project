```python
from google.colab import drive
import warnings
warnings.filterwarnings('ignore')
gym = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/gym_membership.csv')
```

```python
# Gym data
print("\nGym Dataset:")
print(gym.head())
# Display last 5 rows
print("\nLast 5 rows of the dataset:")
print(gym.tail())
# Check data types
print("\nData Types:")
print(gym.dtypes)
print(gym.shape)
print(gym.columns.tolist())
```

```python
from IPython.utils.path import get_ipython_module_path
# Task 1: Basic Matplotlib (3 plots)
# 1. Bar chart of gym membership types
# Use: gym['abonoment_type'].value_counts()
membership_types = gym['abonoment_type'].value_counts()
plt.bar(['Standard', 'Premium'], membership_types.values)
plt.title('Membership Types')
plt.show()

# 2. Histogram of time in gym
# Use: plt.hist(gym['avg_time_in_gym'])
print('\nHistogram of time in gym:')
plt.hist(gym['avg_time_in_gym'], bins=20, color="green")
plt.title("Average time in the gym")
plt.xlabel('Average time')
plt.ylabel("Count")
plt.show()

# 3. Scatter plot: Age vs visits per week
# Use: plt.scatter(gym['Age'], gym['visit_per_week'])
plt.scatter(gym['Age'], gym['visit_per_week'])
plt.xlabel('Age')
plt.ylabel('visit_per_week')
plt.title('Age Vs Visit Per Week')
plt.show()
```

```python
# 1. Box plot: Gender vs Age
# Use: sns.boxplot(data=gym, x='gender', y='Age')
sns.boxplot(data=gym, x='gender', y='Age')
plt.grid(True, which="both", linestyle="--", linewidth=0.5, alpha=0.5)
plt.title('Gender vs Age')
plt.show()
# 2. Count plot: Days per week
# Use: sns.countplot(data=gym, x='days_per_week')
sns.countplot(data=gym, x='days_per_week')
plt.grid(True, which="both", linestyle="--", linewidth=0.5, alpha=0.5)
plt.title('\nDays per week')
plt.show()


# 3. Bar plot: Average visits by membership type
# Use: sns.barplot(data=gym, x='abonoment_type', y='visit_per_week')
sns.barplot(data=gym, x='abonoment_type', y='visit_per_week')
plt.grid(True, which="both", linestyle="--", linewidth=0.5, alpha=0.5)
plt.title('\nAverage visit by membership type')
plt.show()
```
