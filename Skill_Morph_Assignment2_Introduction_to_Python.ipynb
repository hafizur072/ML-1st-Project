```python
courses = ['Accounting','Management', 'Finance', 'Marketing', 'MIS', 'IB']
courses[4]='HRM'
courses.append('OB')
courses.insert(5, "IM")
courses.remove('IB')
print(courses)
popped= courses.pop(-6)
print(popped)
print(courses)
print(len(courses))
print(type(courses))
courses.sort()
print(courses)
courses.reverse()
print(courses)
```

```python
# @title
# Creating a list
fruits = ["apple", "banana", "orange", "mango","pineapple"]
# Accessing elements (indexing starts at 0)


print(fruits[0])       # apple
print(fruits[1])       # banana
# Negative indexing (counts from the end)
print(fruits[-1])      # pineapple
# Slicing lists
print(fruits[1:4])     # ['banana', 'orange’]
# Changing an element
fruits[1] = "grape"
print(fruits)          # ['apple', 'grape', 'orange', 'mango']
```

```python
# @title
#Adding Elements in List
fruits = ["apple", "banana"]
fruits.append("orange")    # Add to end ["apple", "banana", "orange"]
fruits.insert(1, "mango")  # Insert at position
print(fruits)  # ['apple', 'mango', 'banana', 'orange']
```

```python
# @title
#Removing Elements in List
fruits = ["apple", "mango", "banana", "orange"]
fruits.remove("mango")    # Remove by value, updated list: apple, banana, orange
popped = fruits.pop()    # Remove & return last item, apple, banana
print(popped) # orange
print(fruits) # ['apple', 'banana']
```

```python
#organinzing list
fruits = ["orange", "apple", "mango", "banana"]
fruits.sort() # Alphabetical order
print(fruits) # ['apple', 'banana', 'mango', 'orange’]
fruits.reverse() # Reverse order
print(fruits) # ['orange', 'mango', 'banana', 'apple']
```

```python
  # Creating a set
fruits = {"apple", "banana", "orange", "apple"}
print(fruits)  # {'orange', 'banana', 'apple'} (no duplicates)
# Adding elements
fruits.add("mango")
fruits.update(["grape", "kiwi"])
print(fruits)
# Removing elements
fruits.remove("banana")  # Raises error if not found
fruits.discard("pear")
# Set operations
bangladeshi_fruits = {"mango", "jackfruit", "banana"}
imported_fruits = {"apple", "kiwi", "banana"}
# Union (all elements from both sets)
all_fruits = bangladeshi_fruits | imported_fruits
print(all_fruits)
# Intersection (common elements)
common = bangladeshi_fruits & imported_fruits  # {'banana'}
print(common)
```
